```python
#Number of visitors in the Canary islands in time by country of origin (Monthly)
T1 = sns.lineplot(data=df_X, x="Time", y="Tourists", hue='Country');
T1.set_xticklabels(T1.get_xticklabels(), rotation=75, ha="right");
```

```python
#Value of Imports in the Canary islands in time by country of origin (Monthly)
T2 = sns.lineplot(data=df_X, x="Time", y="Imports", hue='Country');
T2.set_xticklabels(T1.get_xticklabels(), rotation=75, ha="right");
```
