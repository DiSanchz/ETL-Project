### Importing libraries and setting up the scrapping tools

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
import time 
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import warnings   
warnings.simplefilter('ignore')

opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    
opciones.add_argument('--start-maximized')         
#opciones.add_argument('user-data-dir=selenium')   
#opciones.add_extension('driver_folder/adblock.crx')       
opciones.add_argument('--incognito')

driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver,options = opciones)
```

### Scrapping

```python
driver = "./chromedriver.exe"
driver = webdriver.Chrome(driver,options = opciones)

############
#Start (Germany)
############

url = 'https://ibestat.caib.es/ibestat/estadistiques/96fc2a7c-fd08-4820-a020-a04f6608f0e7/c775925a-8355-46c3-971a-d228ee26c74c/es/I211001_9001.px'
driver.get(url)

print('Progress 0%')
print('Working on Germany....')

#Selecting data display options for optimal scrapping

ajustes=driver.find_element_by_xpath('//*[@id="seleccionTablaModal"]/img')
ajustes.click()
time.sleep(3)
print('click ajustes')
vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[1]/div[2]/div/div[3]/img[3]')
vaselect.click()
time.sleep(3)
print('vaciando selección (1)')
pais=driver.find_element_by_xpath('//*[@id="busc_2"]')
pais.send_keys('Alemania')
time.sleep(3)
print('select alemania')
lupa=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[1]/div[2]/div/div[4]/img[2]')
lupa.click()
time.sleep(3)
print('click lupa')
vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[2]/div[1]/div/div[3]/img[3]')
vaselect.click()
time.sleep(3)
print('vaciando selección (2)')
valor=driver.find_element_by_xpath('//*[@id="ConfSelCri3"]/option[1]')
valor.click()
time.sleep(3)
print('columna valor')
vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[2]/div[2]/div/div[3]/img[3]')
vaselect.click()
time.sleep(3)
print('vaciando selección (3)')
impor=driver.find_element_by_xpath('//*[@id="ConfSelCri4"]/option[2]')
impor.click()
time.sleep(3)
print('solo imports')
pais=driver.find_element_by_xpath('//*[@id="rows"]/option[2]')
pais.click()
time.sleep(3)
print('país a columna')
flecha=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[4]/div[2]/img[1]')
flecha.click()
time.sleep(3)
print('flecha a columna')
ver=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[5]/a[1]/span')
ver.click()
print('ver la tabla')

#Scrapping

figures = [e.text for e in driver.find_elements_by_class_name('dataCell')]
timetags = [e.text for e in driver.find_elements_by_class_name('izqMed')]

#Creating a Dataframe with the scrapped columns

df_base = pd.DataFrame(list(zip(timetags, figures)),
columns =['Time', 'Alemania'])

display(df_base)


############
#Loop for the other countries' scrapping (Switzerland not present in list)
############

Countries_exc_ger = ['Reino Unido', 'Países Bajos', 'Irlanda', 'Suecia', 'Dinamarca', 'Bélgica', 'Italia', 'Francia', 'República Checa', 'Polonia']
Countries_exc_ger_eng = ['United Kingdom', 'The Netherlands', 'Ireland', 'Sweden', 'Denmark', 'Belgium', 'Italy', 'France', 'Czechia', 'Poland']
#countires' names in the first list are in Spanish as is the language of the source webpage
#the second list is used solely for progress tracking purposes
counter = 0


#Declaring a loop that iterates over the list above and scraps the data for all intended countries
#The structure is the same as in the code above for the first country scrapped (Germany) 
#The main addition is that at the end of each iteration the obtained df is merged to the base one

for i in Countries_exc_ger:
    print(f'Progress {round(((counter + 1)/12)*100,2)}%')
    print(f'Working on {Countries_exc_ger_eng[counter]}....')
    counter += 1

    url = 'https://ibestat.caib.es/ibestat/estadistiques/96fc2a7c-fd08-4820-a020-a04f6608f0e7/c775925a-8355-46c3-971a-d228ee26c74c/es/I211001_9001.px'
    driver.get(url)

    ajustes=driver.find_element_by_xpath('//*[@id="seleccionTablaModal"]/img')
    ajustes.click()
    time.sleep(1)
    vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[1]/div[2]/div/div[3]/img[3]')
    vaselect.click()
    time.sleep(1)
    pais=driver.find_element_by_xpath('//*[@id="busc_2"]')
    pais.send_keys(f'{i}')
    time.sleep(1)
    lupa=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[1]/div[2]/div/div[4]/img[2]')
    lupa.click()
    time.sleep(1)
    vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[2]/div[1]/div/div[3]/img[3]')
    vaselect.click()
    time.sleep(1)
    valor=driver.find_element_by_xpath('//*[@id="ConfSelCri3"]/option[1]')
    valor.click()
    time.sleep(1)
    vaselect=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[2]/div[2]/div/div[3]/img[3]')
    vaselect.click()
    time.sleep(1)
    impor=driver.find_element_by_xpath('//*[@id="ConfSelCri4"]/option[2]')
    impor.click()
    time.sleep(1)
    pais=driver.find_element_by_xpath('//*[@id="rows"]/option[2]')
    pais.click()
    time.sleep(1)
    flecha=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[4]/div[2]/img[1]')
    flecha.click()
    time.sleep(1)
    ver=driver.find_element_by_xpath('//*[@id="contenido_central"]/div[5]/a[1]/span')
    ver.click()

    figures = [e.text for e in driver.find_elements_by_class_name('dataCell')]
    timetags = [e.text for e in driver.find_elements_by_class_name('izqMed')]

#Merging this iteration's DF with the base one
    
    df = pd.DataFrame(list(zip(timetags, figures)),
    columns =['Time', f'{i}'])
    df_base = df.merge(df_base, on='Time')
    display(df_base)
    
#Saving the data into a CSV file after all the data has been scrapped and the loop ends   

print('Progress 99%')    
print('Saving data....')    
df_base.to_csv('balearic_imports_dirty.csv')    
print('Progress 100%')
print('Task accomplished')

#In this particular case the loop is told to display the database with each iteration as due 
#to the relatively moderate size of the "final product" this not unfeasible, but for larger
#scrapping excercises or dataframes this will be totally unpractical.
```
